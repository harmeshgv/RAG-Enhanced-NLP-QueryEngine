```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Standardize all dates to a single format (YYYY-MM-DD)
df["Date"] = df["Date"].dt.strftime("%m-%d-%Y")
```

```python
#On what date did the first transaction occur?

df["Date"].max()
```

```python
df["Product line"] = df["Product line"].str.replace(" and ", " ", regex=False)
```

```python
df.columns = df.columns.str.strip().str.lower()  #.str.replace(r'\s+', '_', regex=True)
```
